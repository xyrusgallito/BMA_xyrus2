```python
import pandas as pd

# Assuming you have combined_df_7 DataFrame with columns ...

# Convert TRANSACTION DATE to datetime if not already
combined_df_7['TRANSACTION DATE'] = pd.to_datetime(combined_df_7['TRANSACTION DATE'])

# Extract quarter and year from TRANSACTION DATE
combined_df_7['Quarter'] = combined_df_7['TRANSACTION DATE'].dt.to_period('Q')
combined_df_7['Year'] = combined_df_7['TRANSACTION DATE'].dt.year

# Group by DIVISION, Quarter, and Year and aggregate TRANSACTION AMOUNT and other features/columns
aggregated_df = combined_df_7.groupby(['DIVISION', 'Quarter', 'Year']).agg({
    'TRANSACTION AMOUNT': 'sum',
    'BATCH-TRANSACTION ID': list,
    'MERCHANT NAME': list,
    'COST CENTRE/WBS ELEMENT/ORDER': list,
    'COST CENTRE/WBS ELEMENT/ORDER DESCRIPTION': list,
    'G/L ACCOUNT': list,
    'G/L ACCOUNT DESCRIPTION': list,
    'MERCHANT TYPE (MCC)': list,
    'MERCHANT TYPE DESCRIPTION': list,
    'PURPOSE': list
    # Add more columns you want to aggregate
}).reset_index()

# Print the aggregated dataframe
aggregated_df.head()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have combined_df_7 DataFrame with columns ...

# Pivot data to create a matrix for clustering ...
pivot_table_normalized = aggregated_df.pivot_table(index='DIVISION', aggfunc=np.mean)

# Determine the optimal number of clusters using the elbow method
inertia_values = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(pivot_table_normalized)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow method graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-cluster sum of squares (WSS)')
plt.title('Elbow Method for Optimal Number of Clusters')

# Find the optimal number of clusters using the "elbow" point method
min_inertia = min(inertia_values[1:])
elbow = inertia_values.index(min_inertia) + 1

# Add a red dot at the optimal number of clusters
plt.scatter(elbow, inertia_values[elbow - 1], color='red', s=100, label='Optimal Cluster')

# Add a dashed line to identify the optimal number of clusters
plt.axvline(x=elbow, color='gray', linestyle='--')

plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
import matplotlib.pyplot as plt

# Assuming you have combined_df_7 DataFrame with columns ...

# Pivot data to create a matrix for clustering ...
pivot_table_normalized = aggregated_df.pivot_table(index='DIVISION', aggfunc=np.mean)

# Calculate Calinski-Harabasz scores for different numbers of clusters
calinski_scores = []
for n_clusters in range(2, 11):  # You can adjust the range as needed
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(pivot_table_normalized)
    calinski_score = calinski_harabasz_score(pivot_table_normalized, cluster_labels)
    calinski_scores.append(calinski_score)

# Plot the Calinski-Harabasz scores
plt.plot(range(2, 11), calinski_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Calinski-Harabasz Score')
plt.title('Calinski-Harabasz Score for Different Numbers of Clusters')
plt.show()

#should be the highest
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have combined_df_7 DataFrame with columns 'DIVISION', 'G/L ACCOUNT DESCRIPTION', 'COST CENTRE/WBS ELEMENT/ORDER'

# Preprocess the data: Convert lists to strings
aggregated_df['COST CENTRE/WBS ELEMENT/ORDER'] = aggregated_df['COST CENTRE/WBS ELEMENT/ORDER'].apply(', '.join)
aggregated_df['G/L ACCOUNT DESCRIPTION'] = aggregated_df['G/L ACCOUNT DESCRIPTION'].apply(', '.join)

# Pivot data to create a matrix for clustering
pivot_table = aggregated_df.pivot_table(index='DIVISION', columns=['COST CENTRE/WBS ELEMENT/ORDER', 'G/L ACCOUNT DESCRIPTION'], values='TRANSACTION AMOUNT', aggfunc='count', fill_value=0)

# Normalize the data (optional)
pivot_table_normalized = pivot_table.div(pivot_table.sum(axis=1), axis=0)

# Perform K-Means clustering
n_clusters = 10  # You can adjust the number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
pivot_table['Cluster'] = kmeans.fit_predict(pivot_table_normalized)

# Calculate cluster counts
cluster_counts = pivot_table['Cluster'].value_counts().sort_index()

# Plotting using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values)
plt.xlabel('Cluster')
plt.ylabel('Number of Divisions')
plt.title('Division Distribution Across Clusters')

# Add total counts on top of each bar
for index, value in enumerate(cluster_counts.values):
    plt.text(index, value + 0.2, str(value), ha='center')

plt.show()
```

```python
# Group divisions by cluster
division_clusters = pivot_table.groupby('Cluster').apply(lambda x: x.index.tolist())

# Print division names per cluster
for cluster, divisions in division_clusters.items():
    print(f"Cluster {cluster}: {divisions}")
```

```python
from sklearn.ensemble import IsolationForest

# Extract relevant features for anomaly detection
X = aggregated_df_2[['TRANSACTION AMOUNT']]

# Create an Isolation Forest model
isolation_forest = IsolationForest(contamination=0.05, random_state=0)  # You can adjust the contamination level

# Fit the model on the data
isolation_forest.fit(X)

# Predict anomalies (1 for normal, -1 for anomaly)
aggregated_df['Anomaly'] = isolation_forest.predict(X)

# Calculate counts of normal and anomaly instances per division
anomaly_counts = aggregated_df.groupby('DIVISION')['Anomaly'].value_counts().unstack(fill_value=0)
anomaly_counts.rename(columns={1: 'Normal', -1: 'Anomaly'}, inplace=True)

# Print the counts
print(anomaly_counts)
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Extract relevant features for anomaly detection
X = aggregated_df_2[['TRANSACTION AMOUNT']]

# Create a Nearest Neighbors model
knn = NearestNeighbors(n_neighbors=5)  # You can adjust the number of neighbors

# Fit the model on the data
knn.fit(X)

# Calculate distances and indices of k-nearest neighbors for each data point
distances, indices = knn.kneighbors(X)

# Calculate an anomaly score based on the distance to the k-th neighbor
k = 5  # You can choose a different value for k
anomaly_scores = distances[:, -1]

# Calculate a threshold for anomaly detection (e.g., using percentiles)
threshold = np.percentile(anomaly_scores, 95)  # You can adjust the percentile

# Identify anomalies based on the threshold
aggregated_df['Anomaly'] = anomaly_scores > threshold

# Print the counts of normal and anomaly instances
print(aggregated_df['Anomaly'].value_counts())
```

The next metric is to include G/L ACCOUNT DESCRIPTION as an added features so see how the division behaves specifically with this feature. Here we can further investigate which product or service they spend more.  

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import IsolationForest

# Extract relevant features for anomaly detection
X = aggregated_df_2[['TRANSACTION AMOUNT']]

# One-hot encode the 'G/L ACCOUNT DESCRIPTION' feature
encoder = OneHotEncoder()
encoded_gl_account = encoder.fit_transform(aggregated_df_2[['G/L ACCOUNT DESCRIPTION']])
encoded_gl_account_df = pd.DataFrame(encoded_gl_account.toarray(), columns=encoder.get_feature_names_out(['G/L ACCOUNT DESCRIPTION']))

# Concatenate the encoded G/L account features with the TRANSACTION AMOUNT feature
X_with_encoded_gl_account = pd.concat([X, encoded_gl_account_df], axis=1)

# Create an Isolation Forest model
isolation_forest = IsolationForest(contamination=0.05, random_state=0)  # You can adjust the contamination level

# Fit the model on the data
isolation_forest.fit(X_with_encoded_gl_account)

# Predict anomalies (1 for normal, -1 for anomaly)
aggregated_df['Anomaly'] = isolation_forest.predict(X_with_encoded_gl_account)

# Calculate counts of normal and anomaly instances per division
anomaly_counts = aggregated_df.groupby('DIVISION')['Anomaly'].value_counts().unstack(fill_value=0)
anomaly_counts.rename(columns={1: 'Normal', -1: 'Anomaly'}, inplace=True)

# Print the counts
print(anomaly_counts)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import LocalOutlierFactor

# Extract relevant features for anomaly detection
X = aggregated_df_2[['TRANSACTION AMOUNT']]

# One-hot encode the 'G/L ACCOUNT DESCRIPTION' feature
encoder = OneHotEncoder()
encoded_gl_account = encoder.fit_transform(aggregated_df_2[['G/L ACCOUNT DESCRIPTION']])
encoded_gl_account_df = pd.DataFrame(encoded_gl_account.toarray(), columns=encoder.get_feature_names_out(['G/L ACCOUNT DESCRIPTION']))

# Concatenate the encoded G/L account features with the TRANSACTION AMOUNT feature
X_with_encoded_gl_account = pd.concat([X, encoded_gl_account_df], axis=1)

# Create a Local Outlier Factor (LOF) model
lof = LocalOutlierFactor(contamination=0.05)  # You can adjust the contamination level

# Fit the model on the data
anomaly_scores = lof.fit_predict(X_with_encoded_gl_account)

# Calculate counts of normal and anomaly instances per division
aggregated_df['Anomaly'] = anomaly_scores
anomaly_counts = aggregated_df.groupby('DIVISION')['Anomaly'].value_counts().unstack(fill_value=0)
anomaly_counts.rename(columns={1: 'Normal', -1: 'Anomaly'}, inplace=True)

# Print the counts
print(anomaly_counts)
```

We can leverage advanced unsupervised machine learning models such as K-Nearest Neighbors (KNN), Local Outlier Factor (LOF), and Isolation Forest to establish a foundation for identifying transactions as either "normal" or "abnormal." These models operate without the need for labeled data, allowing us to uncover potential anomalies within our dataset. It's important to note that refining these models through hyperparameter tuning and thoughtful data preprocessing can significantly enhance their performance.

However, while these models offer a valuable starting point, a thorough investigation is imperative before definitively labeling transactions as anomalous. The expertise of domain-specific professionals is essential to validate and confirm the accuracy of these labels. Anomalies may arise due to a variety of factors such as seasonality, specific events, or unique patterns. Thus, a deeper analysis is necessary to distinguish true anomalies from unusual yet legitimate behaviors.

In essence, the combined power of cutting-edge machine learning models, fine-tuning, and domain knowledge forms a strong basis for classifying transactions as "normal" or "abnormal." But a collaborative approach that includes experts from the relevant field is crucial to ensure that the final labels accurately capture the underlying dynamics of the data.
